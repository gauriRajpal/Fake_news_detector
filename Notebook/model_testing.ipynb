```python
import pandas as pd
```

```python
import numpy as np
```

```python
import joblib
```

```python
model = joblib.load('../models/logistic_model.pkl')
tfidf = joblib.load('../models/tfidf_vectorizer.pkl')
# Loading the model and vectorizer
```

```python
#NOW LOADING TEST DATA
test_df=pd.read_csv("../data/processed/val.csv")
X_test=tfidf.transform(test_df['clean_content'])
Y_test=test_df['label']
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix

y_predict=model.predict(X_test)

print("Test accuracy: ",accuracy_score(Y_test,y_predict))
print("\nClassification Report: ", classification_report(Y_test,y_predict))
print("\nConfusion Matrix: ",confusion_matrix(Y_test,y_predict))
```

```python
import eli5
```

```python
import eli5
from eli5.sklearn import explain_prediction

# Pick one example
idx = 0
sample_text = test_df.iloc[idx]['clean_content']
true_label = test_df.iloc[idx]['label']
predicted_label = model.predict(tfidf.transform([sample_text]))[0]
# Transforms the sample_text into numerical form using the saved tfidf and feeds in the 
# model to get the predicted value

print(f"True Label: {true_label}, Predicted: {predicted_label}")

# Explain prediction
eli5.show_prediction(model, sample_text, vec=tfidf)
# Generates HTML based explanation for the prediction
```
